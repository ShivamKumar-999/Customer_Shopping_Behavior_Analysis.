```python
df["Review Rating"]=df.groupby("Category")["Review Rating"].transform(lambda x: x.fillna(x.median()))
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(" ","_")

df=df.rename(columns={"purchase_amount_(usd)":"purchase_amount"})
     
```

```python
df.columns
```

```python
# Createing a new column age_groups
labels=["Young Adult","Adult","Middle_Age","Senior"]
df["age_group"]=pd.qcut(df["age"],q=4,labels=labels)
```

```python
df[["age","age_group"]].head(15)
```

```python
# Create a column purchase_frequency_days with corrected mappings

frequency_mapping = {
    "Fortnightly": 14,
    "Weekly": 7,  
    "Weely": 7,   
    "Monthly": 30,
    "Quarterly": 90,
    "Bi-Weekly": 14, 
    "Bi-Weely": 14,  
    "Annually": 365,
    "Every 3 Months": 90,  
    "Every 3 Monhts": 90,  
}

# Apply mapping and handle any remaining NaN values
df["purchase_frequency_days"] = df["frequency_of_purchases"].map(frequency_mapping)

# To check for any unmapped values (optional)
unmapped_values = df[df["purchase_frequency_days"].isna()]["frequency_of_purchases"].unique()
if len(unmapped_values) > 0:
    print("Warning: These values were not mapped:", unmapped_values)
    # You could add them to the dictionary and rerun the mapping
```

```python
from sqlalchemy import create_engine
import psycopg2

# Step 1: Connect to PostgreSQL's default database first
username = "postgres"      # default user
password = "Password"      # the password you set during installation
host = "localhost"         # if running locally
port = "5432"              # default PostgreSQL port
default_db = "postgres"    # PostgreSQL's default database

# Connect to default database first
default_engine = create_engine(f"postgresql+psycopg2://{username}:{password}@{host}:{port}/{default_db}")
conn = default_engine.raw_connection()
conn.set_isolation_level(0)  # AUTOCOMMIT mode
cursor = conn.cursor()

# Create the database if it doesn't exist
database = "customer_behavior"
try:
    cursor.execute(f"CREATE DATABASE {database}")
except psycopg2.errors.DuplicateDatabase:
    print(f"Database '{database}' already exists.")
finally:
    cursor.close()
    conn.close()

# Now connect to the newly created database
engine = create_engine(f"postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}")

# Step 2: Load DataFrame into PostgreSQL
# Make sure df is defined before this point
table_name = "customer"   # choose any table name
df.to_sql(table_name, engine, if_exists="replace", index=False)

print(f"Data successfully loaded into table '{table_name}' in database '{database}'.")
```
